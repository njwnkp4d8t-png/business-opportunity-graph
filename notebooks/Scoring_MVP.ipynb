# Scoring MVP (CSV-based)
Simple demo: load exports/blockgroups.csv and compute a basic score.

```python
import pandas as pd, pathlib
BASE = pathlib.Path('..').resolve().parent
EXPORTS = BASE / 'exports'
bg = pd.read_csv(EXPORTS / 'blockgroups.csv')
bg.head()
```

```python
df = bg.copy()
for c in ['medhinc_cy','totpop_cy','n01_bus','s23_emp','n37_sales']:
    if c not in df.columns: df[c] = 0
norm = lambda s: (s - s.min()) / (s.max() - s.min()) if s.max() > s.min() else s*0
df['inc_fit'] = norm(df['medhinc_cy'])
df['pop_fit'] = norm(df['totpop_cy'])
df['proxy_fit'] = norm(df[['n01_bus','s23_emp','n37_sales']].fillna(0).sum(axis=1))
df['score'] = 0.3*df['inc_fit'] + 0.3*df['pop_fit'] + 0.4*df['proxy_fit']
df.sort_values('score', ascending=False).head(20)[['geoid','ctblockgroup','medhinc_cy','totpop_cy','score']]
```
